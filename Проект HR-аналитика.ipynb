# HR-аналитика в компании «Работа с заботой»

**Описание задачи:**

В компании «Работа с заботой» необходимо разработать и обучить модели машинного обучения для предсказания уровня удовлетворённости сотрудников работой и вероятности их увольнения. Это делается для оптимизации управления персоналом, снижения финансовых потерь и уменьшения оттока сотрудников. В рамках проекта нужно построить модель для предсказания уровня удовлетворённости сотрудника работой (0 - совершенно неудовлетворён, 1 - полностью удовлетворён) и модель для предсказания вероятности увольнения сотрудника.

Цель проекта:

Разработать и обучить модели машинного обучения для предсказания уровня удовлетворённости сотрудников работой и вероятности их увольнения на основе предоставленных данных.

Задачи проекта:

Построить модель предсказания уровня удовлетворённости сотрудника: Разработать модель для предсказания уровня удовлетворённости сотрудника работой в компании по шкале от 0 до 1, где 0 — совершенно неудовлетворён, а 1 — полностью удовлетворён.

Построить модель предсказания оттока сотрудников: Разработать модель для предсказания вероятности увольнения сотрудника из компании.

**Этапы работы:**

- Провести предварительный анализ и очистку данных.
- Разделить данные на обучающую и тестовую выборки.
- Разработать и обучить модели машинного обучения.
- Оценить точность и производительность моделей.
- Подготовить отчёт с результатами анализа и предложениями по использованию моделей.

**Описание данных:**

- id — уникальный идентификатор сотрудника;
- dept — отдел, в котором работает сотрудник;
- level — уровень занимаемой должности.
- workload — уровень загруженности сотрудника;
- employment_years — длительность работы в компании (в годах);
- last_year_promo — показывает, было ли повышение за последний год;
- last_year_violations — показывает, нарушал ли сотрудник трудовой договор за последний год;
- supervisor_evaluation — оценка качества работы сотрудника, которую дал руководитель;
- salary — ежемесячная зарплата сотрудника;
- job_satisfaction_rate — уровень удовлетворённости сотрудника работой в компании, целевой признак Задачи 1.
- quit — увольнение сотрудника из компании, целевой признак Задачи 2.

### Шаг 2. Предобработка данных

```python
# Пайплайн для замены пропущенных значений
imputer_pipe = Pipeline(
    [
        (
            'SimpleImputer',
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ),
    ]
)

# Список уникальных значений для каждого столбца (только категориальные признаки)
print('Уникальные значения для каждого категориального столбца:')
[{i: list(job_X_train[i].unique())} for i in job_X_train.columns if job_X_train[i].dtype == 'object']

# Исправление орфографической ошибки в столбце "level"
print('Исправляем орфографическую ошибку в столбце "level":')
job_X_train['level'].replace({'sinior': 'senior'}, inplace=True)

# Проверка уникальных значений в столбце "level" после исправления
print('Проверка уникальных значений в столбце "level" после исправления:')
list(job_X_train['level'].unique())

# Строки с пропущенными значениями в обучающем наборе данных
print('Строки с пропущенными значениями в обучающем наборе данных:')
job_X_train[job_X_train['dept'].isnull() | job_X_train['level'].isnull()]

# Обработка пропущенных значений с использованием пайплайна
nan_cols = ['dept', 'level', 'workload']
job_X_train[nan_cols] = imputer_pipe.fit_transform(job_X_train[nan_cols])

# Проверка на наличие пропущенных значений после обработки
print('Проверка пропущенных значений после обработки:')
display(job_X_train[job_X_train['dept'].isnull() | job_X_train['level'].isnull()])
print(f'Количество пропущенных значений после обработки: \n {job_X_train.isna().sum()}')

# Список уникальных значений для каждого столбца (только категориальные признаки) в тестовом наборе данных
print('Уникальные значения для каждого категориального столбца в тестовом наборе данных:')
[{i: list(job_X_test[i].unique())} for i in job_X_test.columns if job_X_test[i].dtype == 'object']

# Строки с пустыми значениями в столбце "dept" в тестовом наборе данных
print('Строки с пустыми значениями в столбце "dept" в тестовом наборе данных:')
job_X_test.query('dept == " "')

# Строки с пустыми значениями в столбце "workload" в тестовом наборе данных
print('Строки с пустыми значениями в столбце "workload" в тестовом наборе данных:')
job_X_test.query('workload == " "')

# Исправление орфографической ошибки в столбце "level" в тестовом наборе данных
print('Исправляем орфографическую ошибку в столбце "level" в тестовом наборе данных:')
job_X_test['level'].replace({'sinior': 'senior'}, inplace=True)

# Замена пустых значений на NaN и последующая обработка пайплайном
print('Замена пустых значений в столбцах "dept" и "workload" на NaN:')
job_X_test['dept'].replace({' ': np.nan}, inplace=True)
job_X_test['workload'].replace({' ': np.nan}, inplace=True)

# Обработка пропущенных значений с использованием пайплайна в тестовом наборе данных
print('Обработка пропущенных значений в тестовом наборе данных:')
job_X_test[nan_cols] = imputer_pipe.transform(job_X_test[nan_cols])

# Проверка на наличие пропущенных значений после обработки в тестовом наборе данных
print('Проверка пропущенных значений после обработки в тестовом наборе данных:')
display(job_X_test[job_X_test['dept'].isnull() | job_X_test['level'].isnull() | job_X_test['workload'].isnull()])
print(f'Количество пропущенных значений после обработки: \n {job_X_test.isna().sum()}')

# Уникальные значения в столбце "job_satisfaction_rate" целевого тестового набора данных
print('Уникальные значения в столбце "job_satisfaction_rate" в целевом тестовом наборе данных:')
job_y_test['job_satisfaction_rate'].unique()
```

### Шаг 5. Обучение моделей

```python
# Определение функции SMAPE
def smape(y_true, y_pred):
    numerator = np.abs(y_pred - y_true)
    denominator = (np.abs(y_true) + np.abs(y_pred)) / 2
    diff = np.divide(numerator, denominator, out=np.zeros_like(numerator), where=denominator != 0)
    return np.mean(diff) * 100

# Создание пользовательской метрики
smape_score = make_scorer(smape, greater_is_better=False)

# Сетка гиперпараметров для различных моделей
param_grid_test = [[
    {
        'models': [LinearRegression()],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough'],
    }],

    [{
        'models': [DecisionTreeRegressor(random_state=RANDOM_STATE)],
        'models__max_depth': range(1, 21),
        'models__min_samples_leaf': range(1, 6),
        'models__min_samples_split': range(2, 6),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough'],
    }],

    [{
        'models': [LGBMRegressor(random_state=RANDOM_STATE, n_jobs=-1, verbose=-1)],
        'models__max_depth': range(1, 21),
        'models__n_estimators': range(10, 30),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough'],
    }],

    [{
        'models': [RandomForestRegressor(random_state=RANDOM_STATE)],
        'models__max_depth': range(1, 21),
        'models__min_samples_leaf': range(1, 6),
        'models__min_samples_split': range(2, 6),
        'models__n_estimators': range(10, 30),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough'],
    }],

    [{
        'models': [KNeighborsRegressor(n_jobs=-1, weights='distance')],
        'models__n_neighbors': range(5, 20),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough'],
    }],
]

# Пустые списки для записи результатов
models = []
train_score = []
test_score = []

# Шаблон для таблицы результатов
score_table = {
    'model': models,
    'train_SMAPE': train_score,
}

# Перебор комбинаций гиперпараметров для каждой модели и сохранение результатов в `score_table`
best_model = None
best_score = float('inf')

for grid in param_grid_test:
    search = RandomizedSearchCV(
        pipe_final,
        grid,
        n_jobs=-1,
        cv=10,
        n_iter=10,
        scoring=smape_score,
        random_state=RANDOM_STATE
    )

    # Обучение модели
    search.fit(job_X_train_ml, job_y_train_ml)
    current_best_model = search.best_estimator_['models']
    current_train_score = (-1) * search.best_score_

    print(current_best_model)
    print('SMAPE лучшей модели на кросс-валидации по обучающей выборке:', current_train_score)
    print('---')

    # Сохранение результата текущей итерации
    models.append(current_best_model)
    train_score.append(current_train_score)
    
    # Сохранение лучшей модели и ее оценки
    if current_train_score < best_score:
        best_score = current_train_score
        best_model = search.best_estimator_

# Финальное тестирование лучшей модели на тестовой выборке
predictions_test = best_model.predict(job_X_test_ml)
test_score_best_model = smape(job_y_test_ml, predictions_test)
print('Лучшая модель:', best_model)
print('SMAPE лучшей модели на тестовой выборке:', test_score_best_model)

# Создание и вывод таблицы с результатами
score_table['test_SMAPE'] = [None] * len(models)
score_table['test_SMAPE'][models.index(best_model['models'])] = test_score_best_model
scoring_table = pd.DataFrame(score_table)
scoring_table.sort_values(by='train_SMAPE')
```

```python
# Создаем сетку гиперпараметров для моделей
param_grid = [
    {
        'models': [LinearRegression()],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough'],
    },

    {
        'models': [LGBMRegressor(random_state=RANDOM_STATE, n_jobs=-1, verbose=-1)],
        'models__max_depth': range(1, 40),
        'models__n_estimators': range(50, 150),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough'],
    },

]

# Настраиваем RandomizedSearchCV для поиска наилучших гиперпараметров
rs = RandomizedSearchCV(
    pipe_final,
    param_grid,
    n_jobs=-1,
    n_iter=50,
    cv=5,
    scoring=smape_score,
    random_state=RANDOM_STATE
)

# Обучаем модель на тренировочной выборке
rs.fit(job_X_train_ml, job_y_train_ml)

# Выводим лучшую модель и её параметры
print('Лучшая модель и её параметры:\n\n', rs.best_estimator_)
print('---')
print('SMAPE лучшей модели на кросс-валидации по обучающей выборке:', (-1) * rs.best_score_)

# Проверяем модель на тестовой выборке
job_y_test_pred = rs.best_estimator_.predict(job_X_test_ml)
print(f'SMAPE на тестовой выборке: {smape(job_y_test_ml, job_y_test_pred)}')
print('---')
```

```python
# Подготовим данные для SHAP-анализа
model = rs.best_estimator_['models']
preprocessor_pipe = rs.best_estimator_.named_steps['preprocessor']
X_train_scalled = preprocessor_pipe.fit_transform(job_X_train_ml)
X_test_scalled = preprocessor_pipe.transform(job_X_test_ml)
feature_names = rs.best_estimator_[:-1].get_feature_names_out()

# Создаем Explainer
explainer = shap.TreeExplainer(model)

# Получаем SHAP-значения для данных
shap_values = explainer.shap_values(X_test_scalled)

# Визуализиурем важность признаков
display(shap.summary_plot(shap_values, X_test_scalled, feature_names=feature_names))
```

   - Модель предсказания ухода сотрудников помогает компании выявлять потенциальных уходящих сотрудников заранее. Это дает возможность реагировать на ранних стадиях, предлагать персонализированные программы удержания и снижать экономические потери от текучести персонала.

3. **Интерпретируемость моделей:**
   - SHAP-анализ для модели удовлетворённости и анализ важности признаков для модели ухода сотрудников позволяют лучше понимать, какие факторы оказывают наибольшее влияние на решения сотрудников. Это информация может быть использована для разработки целевых программ управления персоналом.

4. **Будущие направления исследований:**
   - Дальнейшие исследования могут включать анализ влияния временных факторов на удовлетворённость и уход сотрудников, использование более сложных моделей для улучшения прогнозов, а также расширение набора данных для учета дополнительных аспектов, таких как климат организационной культуры.

Этот проект демонстрирует значимость анализа данных и применения машинного обучения в управлении персоналом, что может принести значительные выгоды для компаний в улучшении рабочей среды и снижении текучести кадров.
